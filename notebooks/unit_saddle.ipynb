```python
from sts_curves import get_saddle_prob
from tqdm import tqdm
import pickle
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['text.usetex'] = True
plt.rcParams.update({'font.size': 15})
```

```python
unit_prob = ([],[],[]) # rand, fixed, fixed_conj
for n in tqdm(range(1, 251)):

    unit_prob[0].append(get_saddle_prob(n, model='rand'))
    unit_prob[1].append(get_saddle_prob(n, model='fixed'))
    unit_prob[2].append(get_saddle_prob(n, model='fixed_conj'))
```

```python
# pickle.dump(unit_prob, open('unit_prob.sav', 'wb'))
```

```python
up = np.array(unit_prob) / 10000
```

```python
unit_prob[0][18]
```

```python
x = [n for n in range(1, 251)]

plt.scatter(x, up[0], s =.5, label='Random Model')
plt.scatter(x, up[1], s =.4, label='Fixed Model')
plt.scatter(x, up[2], s =.2, label='Fixed-Conjugacy Model')
plt.ylim(.9,1.01)
plt.xlim(1)
plt.legend(loc='lower right')
plt.title('Probabily of Closed Unit Saddle Connection')
plt.ylabel('Proportion')
plt.xlabel('Value of $n$')
```
