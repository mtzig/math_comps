```python
def get_label(s_tuple):
    stratum = ''
    for c in s_tuple:
        stratum += f'{c[0]}'
        if c[1] != 1:
            stratum += f'^{c[1]} '
        else:
            stratum += ' '

    return f'({stratum[:-1]})'
```
