```python
# from brute_stratum import find_strat_sts, test, get_strat_perm
from brute_stratum import *
from collections import Counter
```

```python
def get_label(s_tuple):
    stratum = ''
    for c in s_tuple:
        stratum += f'{c[0]}'
        if c[1] != 1:
            stratum += f'^{c[1]} '
        else:
            stratum += ' '

    return f'({stratum[:-1]})'
```

```python
def pretty_print(vals, conj=False):
    '''
        if conj = True
            prints all permutations up to the same conjugacy class
    '''
    for key in vals:
        print(f'For stratum H{get_label(key)}:')

        conj_perms = set()
        for perms in vals[key]:

            if conj: #note this is only valid if we are looking at fixed h
                conj_p =  tuple((sorted(Counter(get_stratum(perms[1])).items())))
                if conj_p in conj_perms:
                    continue
                conj_perms.add(conj_p)

            h = convert_cycle(perms[0])
            v = convert_cycle(perms[1])
            print(f'    h={h} v={v}')
```
