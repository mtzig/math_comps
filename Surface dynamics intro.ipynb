## There is also a Origami Database that the developers have populated with an exhaustive list of square-tiled surfaces (upto a certain number of squares)

```python
D = OrigamiDatabase()
```

## The database can be 'queried' and asked to give a list of surfaces organized by number of squares, strata etc.

```python
q = D.query(nb_squares=7)
```

```python
q.list()
```

## Here nb_squares is number of squares, and q.list() is a list of surfaces (upto SL2(ZZ) orbit!!) with 5 squares

## Note of caution: due to computational limitations, the list is not exhaustive when the number of squares is large (like 30 or something ..)
